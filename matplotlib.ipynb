```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(0, 10, 1)
b = np.arange(10, 20,1)
a,b
```

```python
plt.plot(a,b, marker='o', linestyle='-', color='green')  # Using keyword arguments for clarity
```

```python
plt.plot(a,b, 'ro-')  # 'ro-' means red color, circle markers, solid line
```

```python
plt.plot(a,b, 'o-', color='blue')  # 'o-' means circle markers, solid line, blue color
```

```python
a = np.arange(1, 10, 1)
b = np.sqrt(a)

plt.plot(a,b, 'ro-')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Plot Title')
```

```python
plt.plot(a,b, color='red', marker='*', linewidth= 4, markersize=15) # 'ro--' means red color, circle markers, dashed line
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Plot Title')
plt.grid(True)  # Adding a grid for better readability
```

```python
a = np.arange(1,10,1)
b = np.arange(11,20,1)
c = np.arange(21,30,1)

plt.Figure(figsize=(10, 5))  # Setting the figure size
plt.subplot(2,2,1)  # Creating a subplot in a 2x2 grid at position 1
plt.plot(a, b, 'ro-')  # Plotting the first subplot 
plt.subplot(2,2,2)  # Creating a subplot in a 2x2 grid at position 2
plt.plot(b, c, 'bo-')  # Plotting the second subplot
plt.subplot(2,2,3)  # Creating a subplot in a 2x2 grid at position 3
plt.plot(c, a, 'go-')  # Plotting the third subplot
plt.subplot(2,2,4)  # Creating a subplot in a 2x2 grid at position 4
plt.plot(a, c, 'yo-')  # Plotting the fourth subplot
plt.tight_layout()  # Adjusting layout to prevent overlap
```

Bar plot

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color='skyblue')  # Creating a bar chart
```

making histogram

```python
# data for histogram
data = np.random.randn(1000)  # Generating random data
plt.hist(data, bins=30, color='purple', alpha=0.7, edgecolor='black')  # Creating a histogram with 30 bins
```

```python
a = np.arange(1, 10, 1)
b = np.arange(11, 20, 1)

plt.scatter(a, b, color='orange', marker='x')  # Creating a scatter plot with orange 'x' markers
```

pie chart

```python
# pie chart data
labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
sizes = [15, 30, 45, 10]  # Sizes of each category
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']  # Colors for each category
explode = (0.1, 0, 0, 0)  # Exploding the first slice for emphasis
plt.pie(sizes, labels=labels,explode=explode, colors=colors, autopct='%1.1f%%', startangle=140)  # Creating a pie chart with percentages
```

Practice on real data

```python
import pandas as pd

df = pd.read_csv('insurance.csv')  # Reading data from a CSV file
df.head()  # Displaying the first few rows of the DataFrame
```

get childern count based on smoking

```python
df.groupby('smoker')['children'].sum().plot(kind='bar', color='lightblue')  # Grouping by 'smoker' and plotting mean 'charges' as a bar chart
plt.xlabel('Smoker Status')  # Setting the x-axis label
plt.ylabel('Total Children')  # Setting the y-axis label
```

```python
res = df.groupby('sex')['smoker'].count().reset_index()  # Group
# res.plot(kind='pie', y='smoker', labels=res)
res
```

```python
s = res.smoker
l = res.sex  # Extracting the 'smoker' column and its labels for pie chart
plt.pie(s, labels=l, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])  # Creating a pie chart for smoker distribution
```
